```python
euro_data.columns = euro_data.columns.swaplevel(0,-1)
euro_data = euro_data.sort_index(axis=1)
```

```python
geo_list = euro_data.columns.get_level_values('GEO').unique().tolist()
countries = [x for x in geo_list if not x.startswith('Euro')]
euro_data = euro_data[countries]
euro_data.columns.get_level_values('GEO').unique()
```

```python
euro_data_f = euro_data.xs(('Percentage of total population', 'Active population'),
                     level=('UNIT', 'INDIC_EM'),
                     axis=1)
euro_data_f.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
euro_data_f = euro_data_f.drop('Total', level='SEX', axis=1)
```

```python
box = euro_data_f.loc['2015'].unstack().reset_index()
sns.boxplot(x="AGE", y=0, hue="SEX", data=box, palette=("husl"), showfliers=False)
plt.xlabel('')
plt.xticks(rotation=35)
plt.ylabel('Percentage of population (%)')
plt.title('Employment in Europe (2015)')
plt.legend(bbox_to_anchor=(1,0.5))
plt.show()
```
